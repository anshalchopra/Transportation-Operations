```python
import requests

start = (-68.3909, -122.5570)   # Toronto
end   = (5.9541, -156.4873)   # Montreal

url = (
    "http://router.project-osrm.org/route/v1/driving/"
    f"{start[1]},{start[0]};{end[1]},{end[0]}"
    "?overview=full&geometries=geojson"
)

response = requests.get(url)
data = response.json()

print(data)

route_coords = data["routes"][0]["geometry"]["coordinates"]
distance_km = data["routes"][0]["distance"] / 1000
duration_hr = data["routes"][0]["duration"] / 3600

import folium

m = folium.Map(location=start, zoom_start=6)

# Convert to lat, lon
route_latlon = [(lat, lon) for lon, lat in route_coords]

folium.PolyLine(
    route_latlon,
    color="blue",
    weight=4,
    tooltip=f"{distance_km:.1f} km, {duration_hr:.1f} hrs"
).add_to(m)

folium.Marker(start, popup="Start").add_to(m)
folium.Marker(end, popup="End").add_to(m)

m.save("route_map.html")

```
